```python
import pandas as pd
import yfinance
import datetime as dt
from pandas_datareader import data as pdr
import yfinance as yf
import warnings
```

```python
end = dt.datetime.now()
start = dt.datetime(2024,1,1)
watch = ["AAPL","META","AMZN","MSFT","NVDA","TSLA","GOOGL"]
df = yf.download(watch, start = start, end = end)
myList = pd.DataFrame()
```

```python
warnings.filterwarnings("ignore")
close = df.Close

def perc(ticker):
    close1 = close[ticker]

    myList[ticker] = close[ticker]

    initial = close[ticker][0]
    
    for i in range(close1.size):
        change = ((close1[i]-initial)/initial)*100
        myList[ticker][i] = change
```

```python
for i in watch:
    perc(i)

myList.plot(title = "Mag7 YTD Percent Returns",ylabel = "% Change")
```
